```python
from pathlib import Path
import cv2
import numpy as np
from ultralytics import YOLO

# Загрузка модели
model = YOLO('yolov8m.pt')

# Конфигурация демо
image_path = str(Path.cwd() / 'images' / 'img1.jpg')
img = cv2.imread(image_path)

# Получение предсказаний
results = model(img, verbose=False)  # Отключение подробного вывода

# Визуализация результатов
annotated_image = results[0].plot()  # Рисуем bounding boxes на изображении

# Сохранение результата
output_path = 'parking_detected.jpg'
cv2.imwrite(output_path, annotated_image)
```
